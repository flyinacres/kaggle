```python
# Import dependencies

import numpy as np
import pandas as pd
```

```python
# Import Matplotlib

import matplotlib.pyplot as plt 
```

```python
%matplotlib inline


x1 = np.linspace(0, 10, 100)


# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--');
```
