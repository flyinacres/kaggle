```python
from itertools import groupby
```

```python
from typing import Any, Iterator
```

```python
numbers: list[[list[int]]] = [[1,2], [3,4,5], [6], [7,8,9]]
sorted_numbers: list[[list[int]]] = sorted(numbers, key=len)

print(sorted_numbers)
```

```python
# groupby is used as a type here, but it is pretty useless (see below, type=type)
grouped_numbers: groupby  = groupby(sorted_numbers, key=len)

print(grouped_numbers)

# Because this is an iterator, if I call 'next' here, when I try to loop 
# through the values (below) it starts after this...
#print(next(grouped_numbers))
```

```python
print(type(groupby))
```

```python
a,b = next(grouped_numbers)
print(a,b)
```

```python
print(a, list(b))
```

```python
for length, group in grouped_numbers:
    print(f'{length}, {list(group)}')
```

```python
numbers2: list[[list[int]]] = [[1,2], [3,4,5], [609], [7,8,9]]
grouped_numbers2: groupby  = groupby(numbers2, key=sum)
```

```python
# Two things: 
# As suspected, there was no need to call sorted() before grouping
# A different key (that applies to the list) works fine

for length, group in grouped_numbers2:
    print(f'{length}, {list(group)}')
```
